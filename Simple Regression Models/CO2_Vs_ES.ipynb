```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
df=pd.read_csv('C:\\Users\\lenovo\\Documents\\Jupyter Projects\\Regression Models\\Sample.csv')
x=df[['ENGINE_SIZE']]
y=df[['CO2_EMISSIONS']]
cdf=df[['ENGINE_SIZE','CO2_EMISSIONS','FUEL_CONSUMPTION*']]
plt.scatter(x,y,color='red',edgecolors='blue',alpha=1,s=40)
plt.xlabel("Engine Size",color='white',fontsize=20)    #Plotting a scatter graph 
plt.ylabel("Co2 Emission",color='white',fontsize=20)
plt.show()
msk=np.random.rand(len(df)) < 0.3
train =df[msk]
test = df[~msk]
plt.scatter(train.ENGINE_SIZE,train.CO2_EMISSIONS)
plt.xlabel("Engine size",color='white',fontsize=20)
plt.ylabel("Emission",color='white',fontsize=20)
plt.show()
from sklearn import linear_model
regr=linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINE_SIZE']])
train_y = np.asanyarray(train[['CO2_EMISSIONS']])
regr.fit(train_x, train_y)
print("The Coffiecient is :", regr.coef_)
print("The intercept is :", regr.intercept_)
plt.scatter(train.ENGINE_SIZE,train.CO2_EMISSIONS)
plt.plot(train_x,regr.coef_[0][0]*train_x+regr.intercept_[0],'-r')
plt.xlabel("Engine Size",color='white',size=20)
plt.ylabel("Emission",color='white',size=20)
plt.show()
from sklearn.metrics import r2_score
test_x=np.asanyarray(test[['ENGINE_SIZE']])
test_y=np.asanyarray(test[['CO2_EMISSIONS']])
test_y_hat=regr.predict(test_x)
z=np.mean(np.absolute((test_y_hat - test_y)))
print("Mean Absolute Error (MAE): %.2f " % np.mean(np.absolute((test_y_hat - test_y))))
print("The Mean Sum of Square Error is : %.2f " % np.mean((test_y_hat-test_y)**2))
print("The R2 Score is : %.2f " % r2_score(test_y_hat,test_y))
z=z/100
print("To further Predict the value of CO2 Emission made by a certain of engine size we could use the above model")
print("for that please put the Engine size")
ES=float(input("Please put the Engine Size "))
predicted_value=regr.coef_[0][0]*ES +regr.intercept_
print("The Predicted Value of CO2 emission is ", predicted_value)
```
