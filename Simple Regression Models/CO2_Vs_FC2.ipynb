```python
df=pd.read_csv('C:\\Users\\lenovo\\Documents\\Jupyter Projects\\Regression Models\\Sample.csv')
df.head(10)
cdf1=df[['Unnamed: 9']]
```

```python
plt.scatter(cdf1,df.CO2_EMISSIONS)
plt.ylabel("CO2_EMISSIONS",color='white',size=20)
plt.xlabel("Fuel Consumption 2",color='white',size=20)
plt.show()
```

```python
msk=np.random.rand(len(df))<0.3
train=df[msk]
test=df[~msk]
```

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
train_x=np.asanyarray(train[['Unnamed: 9']])
cdf=train_x
train_y=np.asanyarray(train[['CO2_EMISSIONS']])
regr.fit(train_x,train_y)
print("The coefficient is: ", regr.coef_)
print("The Intercept is : ",regr.intercept_)
```

```python
plt.scatter(cdf,train.CO2_EMISSIONS)
plt.ylabel("Trained CO2 EMISSIONS",color='white',size=20)
plt.xlabel("Trained Fuel Consumption 2",color='white',size=20)
plt.plot(train_x,train_x*regr.coef_[0][0]+regr.intercept_[0],'-r')
plt.show()
```

```python
from sklearn.metrics import r2_score
test_x=np.asanyarray(test[['Unnamed: 9']])
test_y=np.asanyarray(test[['CO2_EMISSIONS']])
test_y_hat=regr.predict(test_x)
print('The MAE is : %.2f'%np.mean(np.absolute(test_y_hat-test_y)))
print("The MSE is : %.2f"%np.mean((test_y_hat-test_y)**2))
print("The R2 Score is : %.2f" %r2_score(test_y_hat,test_y))
```
